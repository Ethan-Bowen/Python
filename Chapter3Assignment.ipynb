```python
seconds_per_minute = 60
seconds_per_hour = 60
hours_per_day = 24

seconds_per_hour = seconds_per_minute * seconds_per_hour

print(f"There are {seconds_per_hour} seconds in an hour.")

seconds_per_day = seconds_per_hour * hours_per_day

print(f"There are {seconds_per_day} seconds in a day.")

float_div_answer = seconds_per_day / seconds_per_hour

int_div_answer = seconds_per_day // seconds_per_hour

print(f"The float value of division is {float_div_answer}.")
print(f"The integer value of division is {int_div_answer}.")
```
