```python
#Ethan Bowen
#Eligibility Test
#This program check's a student's eligibility for the Dean's List or the Honor Roll

while True:
    #This asks for the student's last name. The .strip is to remove any spaces at the front and end of the input
    #to ensure it doesn't make the final output look awkward. 
    last_name = input("Enter the student's last name (or 'ZZZ' to quit): ").strip()
    if last_name.upper() == 'ZZZ':
        print("Exiting the program. Goodbye!")
        break

    #This asks for the student's first name. The .strip is here for the same reason as the last name input.
    first_name = input("Enter the student's first name: ").strip()

    #This asks for the student's GPA.
    try:
        gpa = float(input("Enter the student's GPA: "))
    except ValueError:
        print("Invalid GPA. Please enter a numeric value.")
        continue

    #This checks the qualifications of the student
    if gpa >= 3.5:
        print(f"{first_name} {last_name} qualifies for the Dean's List.")
    elif gpa >= 3.25:
        print(f"{first_name} {last_name} qualifies for the Honor Roll.")
    else:
        print(f"{first_name} {last_name} does not qualify for the Dean's List or Honor Roll.")
```
