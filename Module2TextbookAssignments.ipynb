```python
#Ethan Bowen
#4.2

#Assigns true and/or false to the variables. Default is true/true
small = True
green = True

#This is the list of items to check
items = ["cherry", "pea", "watermelon", "pumpkin"]

#This checks each item and sees if they match the criteria above
for item in items:
    if item == "cherry":
        if small and not green:
            print(f"{item} matches: small and not green.")
        else:
            print(f"{item} does not match.")
    elif item == "pea":
        if small and green:
            print(f"{item} matches: small and green.")
        else:
            print(f"{item} does not match.")
    elif item == "watermelon":
        if not small and green:
            print(f"{item} matches: not small and green.")
        else:
            print(f"{item} does not match.")
    elif item == "pumpkin":
        if not small and not green:
            print(f"{item} matches: not small and not green.")
        else:
            print(f"{item} does not match.")
```
