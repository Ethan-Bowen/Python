```python
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def __str__(self):
        return f"Vehicle Type: {self.vehicle_type}"


class Automobile(Vehicle):
    def __init__(self, vehicle_type, year, make, model, doors, roof):
        super().__init__(vehicle_type)
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

    def __str__(self):
        return (
            f"{super().__str__()}\n"
            f"Year: {self.year}\n"
            f"Make: {self.make}\n"
            f"Model: {self.model}\n"
            f"Number of Doors: {self.doors}\n"
            f"Roof Type: {self.roof}"
        )


def get_int_input(prompt, valid_values=None):
    while True:
        try:
            value = int(input(prompt))
            if valid_values and value not in valid_values:
                print(f"Please enter one of the following: {valid_values}")
                continue
            return value
        except ValueError:
            print("Invalid input. Please enter a number.")


if __name__ == "__main__":
    print("=== Vehicle Information Input ===")

    vehicle_type = input("Enter vehicle type (e.g., Car, Truck, SUV): ").strip()
    year = get_int_input("Enter year (e.g., 2022): ")
    make = input("Enter make (e.g., Toyota): ").strip()
    model = input("Enter model (e.g., Corolla): ").strip()
    doors = get_int_input("Enter number of doors (2 or 4): ", valid_values=[2, 4])
    roof = input("Enter roof type (e.g., Solid, Sunroof): ").strip()

    # Create Automobile object
    my_auto = Automobile(vehicle_type, year, make, model, doors, roof)

    print("\n=== Vehicle Details ===")
    print(my_auto)
```
