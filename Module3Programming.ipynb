```python
list = ["mozzarella", "cinderella", "salmonella"]
if len(list) >= 1 and list[0]:
    list[0] = list[0].upper()
if len(list) >= 2 and list[1]:
    list[1] = list[1][0].upper() + list[1][1:]
if len(list) >= 3 and list[2]:
    list.pop()

print(list)
```

```python
def get_odds():
    for number in range(10):
        if number % 2 == 1:
            yield number
if __name__ == "__main__":
    count = 0
    for odd in get_odds():
        count += 1
        if count == 3:
            print(f"The third odd number is: {odd}")
            break
```
