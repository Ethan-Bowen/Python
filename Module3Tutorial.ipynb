```python
class Solution:
    def sort012(self, arr):

        low, mid, high = 0, 0, len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            elif arr[mid] == 2:
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1
            else:
                raise ValueError("Array contains elements other than 0, 1, or 2.")

        return arr


if __name__ == "__main__":
    try:

        user_input = input("Enter array elements (only 0, 1, 2) separated by spaces: ").strip()
        arr = list(map(int, user_input.split()))


        if not all(x in (0, 1, 2) for x in arr):
            raise ValueError("Input must contain only integers 0, 1, or 2.")


        sol = Solution()
        sorted_arr = sol.sort012(arr)

        print(sorted_arr)

    except ValueError as ve:
        print("Error:", ve)
    except Exception as e:
        print("Unexpected error:", e)
```

```python
class Solution:
    def binary_search(self, arr, k):

        left, right = 0, len(arr) - 1
        
        while left <= right:
            mid = (left + right) // 2
            if arr[mid] == k:
                return mid
            elif arr[mid] < k:
                left = mid + 1
            else:
                right = mid - 1
        
        return -1

if __name__ == "__main__":
    try:
        arr_input = input("Enter sorted integers separated by spaces: ").strip()
        if not arr_input:
            raise ValueError("Array cannot be empty.")
        
        arr = list(map(int, arr_input.split()))
        
        if arr != sorted(arr):
            raise ValueError("The array must be sorted in ascending order.")
        
        # Get integer k from user
        k_input = input("Enter the integer to search for: ").strip()
        if not k_input.lstrip('-').isdigit():
            raise ValueError("Please enter a valid integer for k.")
        
        k = int(k_input)
        
        solution = Solution()
        result = solution.binary_search(arr, k)
        
        print(result)
        
    except ValueError as e:
        print(f"Error: {e}")
```
